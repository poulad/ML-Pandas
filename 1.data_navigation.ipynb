# Data Navigation

```python
import pandas
data = pandas.read_csv('data/train.csv')
```

```python
data.head()
```

```python
# locating based on integer position
data.iloc[2]  # third row
```

```python
# show a range
data.iloc[3:9]  # rows 4 to 9
```

```python
# Show labels for the first 5 rows
data.loc[:5, ['Name', 'Sex', "Age"]]
```

```python
# apply a condition
data.loc[data['Ticket'] == "STON/O2. 3101282"]
```

```python
data.groupby(['Sex', "Survived"])['PassengerId'].count()
```

```python
passenger_class_grouping = data.groupby('Pclass')
```

```python
passenger_class_grouping["Age"].mean()
```

```python
passenger_class_grouping["Fare"].mean()
```
